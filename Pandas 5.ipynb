## # `.apply`
`.apply` is both a **Series and **DataFrame** method. 

### `.apply` on `Series`
As a **`Series`** method, `apply` allows you to **apply** a function to each element in a series.

```python
df.head()
```

```python
df['Name']
```

```python
df['Name'].str.split(',')
```

```python
def get_lname(lst, pos):
    return(lst[pos])
```

```python
df['Name'].str.split(',').apply(get_lname, pos=0)
```

```python
def status(x):
    '''The function returns whether a person is a Teenager, Youth or an Adult'''
    if x > 0 and x <= 18:
        return('Teenager')
    elif x > 18 and x <=40:
        return('Youth')
    elif x > 40:
        return('Adult')
    else:
        return('Others')
```

```python
status(53)
```

```python
df['Status']  = df['Age'].apply(status)
```
