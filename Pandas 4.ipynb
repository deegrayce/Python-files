# How to Use Pandas `Query` : `df.query()`

```python
import pandas as pd
```

```python
df = pd.read_excel('Orders Data.xlsx')
df
```

```python
df.dtypes
```

```python
# Select alll record/rows where Quantity is greater than 150
df.query('Quantity > 150')
```

```python
df[df['Quantity'] > 150]
```

```python
df.query('Product == "Noodles"')
```

```python
df[df['Product'] == 'Noodles']
```

```python
df[(df['Product'] == 'Noodles') & (df['Quantity'] > 150)]
```

```python
df.query('(Product == "Noodles") and (Quantity > 150)')
```

```python
df.query('(Product == "Noodles") or (Quantity > 150)')
```
