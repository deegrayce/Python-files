# Dealing with Categorical data

```python
import pandas as pd
```

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df1 = df.copy()
```

## Dummy variables

### Using `.map`

```python
df1['Sex'] = df1['Sex'].map({'male':1, 'female':0})
```

```python
df1.sample(20)
```

```python
df1['Embarked'].nunique()
```

```python
df1['Embarked'].unique()
```

```python
df1['Embarked'] = df1['Embarked'].map({'S':1, 'C':2, 'Q':3})
```

```python
df1.sample(20)
```

### `.get_dummies`

```python
df2 = df.copy()
```

```python
pd.get_dummies(df2['Sex'], drop_first=True)
```

```python
pd.get_dummies(df2['Embarked'], drop_first=True)
```
