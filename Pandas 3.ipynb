### #: Filtering DataFrame based on Multiple conditions

We can filter based on multiple conditions using the **and** `&` or the **or** `|` logical operator

````python
df[(df['Product'] == 'Noodles') & (df['Quantity'] > 150)]

#['Product'] == 'Noodles'

#df['Quantity'] > 150
````

````python
df[(df['Product'] == 'Noodles') | (df['Quantity'] > 150)]
````

## Filter DataFrame Rows based on List of Values

````python
df[(df['Product'] == 'Deodorant') | (df['Product'] == 'Soft Drinks') | (df['Product'] == 'Coffee')]
````
